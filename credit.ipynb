```python
def status(s):
    if s == 'C':
        return 0
    elif s == 'X':
        return 0
    elif s == '0':
        return 0
    elif s == '1':
        return 0
    elif s == '2':
        return 1
    elif s == '3':
        return 1
    elif s == '4':
        return 1
    elif s == '5':
        return 1
```

```python
dataset["STATUS"] = dataset['STATUS'].apply(lambda x:status(x))
```

```python
dataset["STATUS"]
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols = dataset.select_dtypes(include = "object").columns #gives a list of the columns with type 'object' in the dataframe
cols
```

```python
le = LabelEncoder()

for feature in cols:
    dataset[feature] = le.fit_transform(dataset[feature].astype(str))
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

dataset[['AMT_INCOME_TOTAL','DAYS_EMPLOYED','MONTHS_BALANCE']] = scaler.fit_transform(dataset[['AMT_INCOME_TOTAL','DAYS_EMPLOYED','MONTHS_BALANCE']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
# x_train, x_test, y_train, y_test = train_test_split(x, y)

x = dataset[['CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'NAME_INCOME_TYPE','NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE','AMT_INCOME_TOTAL','DAYS_EMPLOYED','MONTHS_BALANCE']].values

y = dataset['STATUS'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)

pred = model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
score = np.sqrt(mean_absolute_error(pred, y_test))
```
